```python
import pandas as pd

df = pd.read_csv("./work/airthings_history.csv", delimiter=";")

df
```

```python
# Convert 'recorded' column to datetime
df['recorded'] = pd.to_datetime(df['recorded'], format="mixed")

# Set 'recorded' as index
df.set_index('recorded', inplace=True)

# Resample to 5 minute intervals, aggregating with mean
df_5min = df.resample('5T').mean()

# Drop rows with any NaNs
df_5min_clean = df_5min.dropna()

df_5min_clean
```
